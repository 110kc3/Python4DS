Winds are stronger during the night and at mornings (dir W)

```python
plt.xlabel('hour')
plt.ylabel('Degrees')
plt.title('Wind Direction  [10 m above gnd]')
plt.grid()
plt.plot(df_c['wDir'],'ro')
```

from 11:00 to 18:00 in general wind direction is "W" (300o). Other hours is toward NorthEast (NE) - (50o)

```python
def plotWindPoint(pnt):
    ws = df_c.loc[pnt,'wSpeed']
    wd = df_c.loc[pnt,'wDir']      # wind direction in degrees
    print("Sample:", pnt, "  Wind Speed:", ws, "  Wind Direction:", wd)
    fig = plt.figure(figsize=(10, 10))
    ax = plt.subplot(111, projection = 'polar')
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.margins(0)
    theta = wd*np.pi/180
    ax.plot([0, 0], [0, 30], 'ow') # Max Range of wind speed
    ax.plot([0, theta], [0, ws], linewidth=4)
    
```

```python
plotWindPoint(191)
```

```python
def plotWindRep(pnt, rpt=5):
    fig = plt.figure(figsize=(10, 10))
    ax = plt.subplot(111, projection = 'polar')
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.margins(0)

    ax.plot([0, 0], [0, 30], 'ow')
    for i in range (0, rpt):
        ws = df_c.loc[(pnt-i),'wSpeed']
        wd = df_c.loc[pnt-i,'wDir']      # wind direction in degrees
        theta = wd*np.pi/180
        #print("Sample:", (pnt-i), "  Wind Speed:", ws, "  Wind Direction:", wd)
        ax.plot([0, theta], [0, ws], linewidth=4)
    
```

```python
plotWindRep(191, 190)
```

```python
def plotWindRange(pnt1, pnt2=5):
    fig = plt.figure(figsize=(10, 10))
    ax = plt.subplot(111, projection = 'polar')
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.margins(0)

    ax.plot([0, 0], [0, 30], 'ow')
    for i in range (pnt1, pnt2):
        ws = df_c.loc[i,'wSpeed']
        wd = df_c.loc[i,'wDir']      # wind direction in degrees
        theta = wd*np.pi/180
        #print("Sample:", i, "  Wind Speed:", ws, "  Wind Direction:", wd)
        ax.plot([0, theta], [0, ws], linewidth=4)
    
```

```python
plotWindRange(19, 23)
```
